# **Exploratory Analysis**

# **Gender**

```python
ax = sns.countplot(x='Gender',data=data)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.groupby(['Gender'], as_index=False)['Salary'].sum().sort_values(by='Salary', ascending=False)
```

```python
Salary_gen=data.groupby(['Gender'], as_index=False)['Salary'].sum().sort_values(by='Salary', ascending=False)

sns.barplot(x='Gender',y='Salary',data=Salary_gen)
```

# From the above graphs we can see that Male employees are more than female employees and they both get almost equal amount of salary


## **Experience(Years)**

```python
data.columns
```

```python
ax=sns.countplot(data=data, x='Experience (Years)',hue='Gender')

for bars in ax.containers:
   ax.bar_label(bars)
   plt.show()
```

```python
# Total salary vs Experience

Salary_gen = data.groupby(['Experience (Years)'],as_index=False)['Salary'].sum().sort_values(by='Salary',ascending=False)

sns.barplot(x='Experience (Years)',y='Salary',data=Salary_gen)
plt.show()
```

## From the above graphs we can see that there are more experinced male than female in this company and salary gets fluctuated according to the experience

# **Position**

```python
sns.set(rc={'figure.figsize':(30,5)})
ax=sns.countplot(data=data,x='Position')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
Salary_gen= data.groupby(['Position'],as_index=False)['Salary'].sum().sort_values(by='Salary',ascending=False)
sns.set(rc={'figure.figsize':(30,5)})
sns.barplot(data = Salary_gen,x='Position',y='Salary')
plt.show()
```

# From the above graphs we can see that maximum number of employees are at Web Development position. IT Manager position gets the highest salary.

# **CONCLUSION**

### **In this company there are maximum male employees which are experienced as well.Here, maximum number of employees are there in the Web Development position but IT Manager position gets highest salary.**
